```python
def create_feed_url_list (after_str: str, before_str: str) -> list:
    urls = []
    date_format = "%Y-%m-%d"
    
    after_date = datetime.strptime(after_str, date_format)
    before_date = datetime.strptime(before_str, date_format)
    next_date = after_date + timedelta(days=1)
   
    while next_date <= before_date:     
        after = after_date.strftime("%Y-%m-%d")
        before = next_date.strftime("%Y-%m-%d")
        urls.append(f'https://news.google.com/rss/search?q={query}+before:{before}+after:{after}&hl={language}&gl={country}&ceid={country}:{language}')
        after_date = next_date
        next_date = next_date + timedelta(days=1)

    return urls
```

```python
# Retrieve RSS feed
with open(output_file, "w", newline="") as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(["RSS URL", "Title", "Link", "Published", "Description", "Source"])
    for url in create_feed_url_list(after, before):
        feed = feedparser.parse(url)
        if feed.entries:
            for entry in feed.entries:
                rss_url = url
                title = entry.title
                link = entry.link
                description = entry.description
                pubdate = entry.published
                source = entry.source
                output_data.append([rss_url, title, link, pubdate, description, source])
                print(f"Title: {title}\nLink: {link}\nDescription: {description}\nPublished: {pubdate}\nSource: {source}")
                print("--")
        else:
            print("Nothing Found!")

        with open(output_file, "w", newline="") as csvfile:
            csv_writer = csv.writer(csvfile)
            csv_writer.writerow(["RSS URL", "Title", "Link", "Published", "Description", "Source"])
            for row in output_data:
                csv_writer.writerow(row)
```
